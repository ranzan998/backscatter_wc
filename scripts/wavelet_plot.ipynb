```python
import matplotlib.pyplot as plt
import xarray as xr
import numpy as np
import matplotlib.dates as mdates
import pandas as pd
import os
```

# for 40 m wavelet ananlysis plot

```python
path = "/media/scilab/disk_ranjan/works/westcoast_adcp1/backscatternew/8process/wavelet/40m_wavelet/"
files = sorted([f for f in os.listdir(path) if f.endswith(".nc")])
```

```python
files
```

```python
ok = xr.open_dataset(path+files[0])
mu = xr.open_dataset(path+files[1])
jg = xr.open_dataset(path+files[2])
go = xr.open_dataset(path+files[3])
ud = xr.open_dataset(path+files[4])
ko = xr.open_dataset(path+files[5])
ka = xr.open_dataset(path+files[6])

customticks = [15*24,30*24,90*24,180*24,360*24]
customticks = [np.log2(e) for e in customticks]
customlabel = ["15","30","90","180","360"]
names = ["Okha",'Mumbai','Jaigarh',"Goa","Udupi","Kollam","Kanyakumari"]
figlab = ["a)","b)","c)","d)","e)","f)","g)"]
```

```python
ds = [xr.open_dataset(path+f) for f in files]
```

```python

fig, ax = plt.subplots(nrows=7,ncols=1,figsize=(8,10),sharey=True,sharex=True)
plt.subplots_adjust(left=0.08,right=0.96,top=0.945,bottom=0.04)

# OKHA
figcol = ax[0].contourf(ok.time,np.log2(ok.period),np.log2(ok.wavpow1),extend='both',cmap="turbo",levels=np.arange(10,27,1))
ax[0].plot(ok.time, np.log2(ok.coi1),color='k',linewidth=0.7)
ax[0].fill_between(ok.time, np.log2(ok.coi1),16,color='k',alpha=0.2,hatch='xxx')
extent = [ok.time.min(), ok.time.max(), 0, max(ok.period)]
ax[0].contour(ok.time, np.log2(ok.period), ok.sig1,levels=[-99,1], colors='k', linewidths=0.5,extent=extent)

ax[0].contourf(ok.time,np.log2(ok.period),np.log2(ok.wavpow2),extend='both',cmap="turbo",levels=np.arange(10,27,1))
ax[0].plot(ok.time, np.log2(ok.coi2),color='k',linewidth=0.7)
ax[0].fill_between(ok.time, np.log2(ok.coi2),16,color='k',alpha=0.2,hatch='xxx')
extent = [ok.time.min(), ok.time.max(), 0, max(ok.period)]
ax[0].contour(ok.time, np.log2(ok.period), ok.sig2,levels=[-99,1], colors='k', linewidths=0.5,extent=extent)


# mumbai
ax[1].contourf(mu.time,np.log2(mu.period),np.log2(mu.wavpow1),extend='both',cmap="turbo",levels=np.arange(10,27,1))
ax[1].plot(mu.time, np.log2(mu.coi1),color='k',linewidth=0.7)
ax[1].fill_between(mu.time, np.log2(mu.coi1),16,color='k',alpha=0.2,hatch='xxx')
extent = [mu.time.min(), mu.time.max(), 0, max(mu.period)]
ax[1].contour(mu.time, np.log2(mu.period), mu.sig1,levels=[-99,1], colors='k', linewidths=0.5,extent=extent)

# Jaigarh
ax[2].contourf(jg.time,np.log2(jg.period),np.log2(jg.wavpow1),extend='both',cmap="turbo",levels=np.arange(10,27,1))
ax[2].plot(jg.time, np.log2(jg.coi1),color='k',linewidth=0.7)
ax[2].fill_between(jg.time, np.log2(jg.coi1),16,color='k',alpha=0.2,hatch='xxx')
extent = [jg.time.min(), jg.time.max(), 0, max(jg.period)]
ax[2].contour(jg.time, np.log2(jg.period), jg.sig1,levels=[-99,1], colors='k', linewidths=0.5,extent=extent)

ax[2].contourf(jg.time,np.log2(jg.period),np.log2(jg.wavpow2),extend='both',cmap="turbo",levels=np.arange(10,27,1))
ax[2].plot(jg.time, np.log2(jg.coi2),color='k',linewidth=0.7)
ax[2].fill_between(jg.time, np.log2(jg.coi2),16,color='k',alpha=0.2,hatch='xxx')
extent = [jg.time.min(), jg.time.max(), 0, max(jg.period)]
ax[2].contour(jg.time, np.log2(jg.period), jg.sig2,levels=[-99,1], colors='k', linewidths=0.5,extent=extent)

"""
ax[2].contourf(jg.time,np.log2(jg.period),np.log2(jg.wavpow3),extend='both',cmap="turbo",levels=np.arange(10,27,1))
ax[2].plot(jg.time, np.log2(jg.coi3),color='k',linewidth=0.7)
ax[2].fill_between(jg.time, np.log2(jg.coi3),16,color='k',alpha=0.2,hatch='xxx')
extent = [jg.time.min(), jg.time.max(), 0, max(jg.period)]
ax[2].contour(jg.time, np.log2(jg.period), jg.sig3,levels=[-99,1], colors='k', linewidths=0.5,extent=extent)
"""

# goa
ax[3].contourf(go.time,np.log2(go.period),np.log2(go.wavpow1),extend='both',cmap="turbo",levels=np.arange(10,27,1))
ax[3].plot(go.time, np.log2(go.coi1),color='k',linewidth=0.7)
ax[3].fill_between(go.time, np.log2(go.coi1),16,color='k',alpha=0.2,hatch='xxx')
extent = [go.time.min(), go.time.max(), 0, max(go.period)]
ax[3].contour(go.time, np.log2(go.period), go.sig1,levels=[-99,1], colors='k', linewidths=0.5,extent=extent)

#Udupi
ax[4].contourf(ud.time,np.log2(ud.period),np.log2(ud.wavpow1),extend='both',cmap="turbo",levels=np.arange(10,27,1))
ax[4].plot(ud.time, np.log2(ud.coi1),color='k',linewidth=0.7)
ax[4].fill_between(ud.time, np.log2(ud.coi1),16,color='k',alpha=0.2,hatch='xxx')
extent = [ud.time.min(), ud.time.max(), 0, max(ud.period)]
ax[4].contour(ud.time, np.log2(ud.period), ud.sig1,levels=[-99,1], colors='k', linewidths=0.5,extent=extent)

ax[4].contourf(ud.time,np.log2(ud.period),np.log2(ud.wavpow2),extend='both',cmap="turbo",levels=np.arange(10,27,1))
ax[4].plot(ud.time, np.log2(ud.coi2),color='k',linewidth=0.7)
ax[4].fill_between(ud.time, np.log2(ud.coi2),16,color='k',alpha=0.2,hatch='xxx')
extent = [ud.time.min(), ud.time.max(), 0, max(ud.period)]
ax[4].contour(ud.time, np.log2(ud.period), ud.sig2,levels=[-99,1], colors='k', linewidths=0.5,extent=extent)

# Kollam
ax[5].contourf(ko.time,np.log2(ko.period),np.log2(ko.wavpow1),extend='both',cmap="turbo",levels=np.arange(10,27,1))
ax[5].plot(ko.time, np.log2(ko.coi1),color='k',linewidth=0.7)
ax[5].fill_between(ko.time, np.log2(ko.coi1),16,color='k',alpha=0.2,hatch='xxx')
extent = [ko.time.min(), ko.time.max(), 0, max(ko.period)]
ax[5].contour(ko.time, np.log2(ko.period), ko.sig1,levels=[-99,1], colors='k', linewidths=0.5,extent=extent)

#kanyakumari
ax[6].contourf(ka.time,np.log2(ka.period),np.log2(ka.wavpow1),extend='both',cmap="turbo",levels=np.arange(10,27,1))
ax[6].plot(ka.time, np.log2(ka.coi1),color='k',linewidth=0.7)
ax[6].fill_between(ka.time, np.log2(ka.coi1),16,color='k',alpha=0.2,hatch='xxx')
extent = [ka.time.min(), ka.time.max(), 0, max(ka.period)]
ax[6].contour(ka.time, np.log2(ka.period), ka.sig1,levels=[-99,1], colors='k', linewidths=0.5,extent=extent)

for i in range(7):
    ax[i].axhline(np.log2(24*360),color='w',linestyle='--',linewidth=1.4)
#    ax[i].axhline(np.log2(24*90),color='w',linestyle='--',linewidth=1.4)
    ax[i].axhline(np.log2(24*180),color='w',linestyle='--',linewidth=1.4)
    ax[i].set_yticks(customticks)
    ax[i].set_yticklabels(customlabel)
    ax[i].text(0.010, 0.97, names[i],transform=ax[i].transAxes, fontsize=9, va='top',
            ha='left',color='black',fontweight="bold",bbox=dict(facecolor='white', edgecolor='black', boxstyle='round', alpha=1))    
    
    dsm = ds[i].resample(time='ME').mean()
    tklb = []
    for t in dsm.time.values:
        monlb = str(pd.to_datetime(t).month_name()[0:3])
        if monlb =="Jun":
            yrlb = str(pd.to_datetime(t).year)
            tklb.append(f'{monlb[0]}\n {yrlb}')
        elif monlb == "Sep":
            tklb.append(f'{monlb[0]}')
        elif monlb == "Dec":
            tklb.append(f'{monlb[0]}')
            ax[i].axvline(t,color='w',linewidth=0.6)
        elif monlb == "Mar":
            tklb.append(f'{monlb[0]}')
        else: 
            tklb.append("")
    ax[0].set_xticks(dsm.time,tklb)      
    ax[i].set_xlim(xmin=np.array('2017-10-01T00:00:00', dtype='datetime64[ns]'),xmax=np.array('2023-12-20T11:00:00', dtype='datetime64[ns]'))
ax[3].set_ylabel("period (days)")
ax[0].set_ylim(ymin=7,ymax=14)

cax = fig.add_axes([0.15, 0.98, 0.75,0.0125])
cbar = fig.colorbar(figcol, cax=cax, orientation="horizontal",)
cbar.ax.tick_params(labelsize=10,)

plt.savefig(path+"west_coast_wavelet_40m_scale.png",dpi=400)
plt.savefig("/media/scilab/disk_ranjan/works/backscatter_wc/figures/west_coast_wavelet_40m_scale.png",dpi=400)
```
